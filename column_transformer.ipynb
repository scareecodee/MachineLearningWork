```python
import pandas as pd
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
# seperate input and output variable
x=df[["age","gender","fever","cough","city"]]
y=df[["has_covid"]]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2
)
```

```python
x_train
```

```python
x_test
```

```python
#column tranformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(
    transformers=[
    ("scaler",StandardScaler(),["age"]),
    ("SimpleImputer",SimpleImputer(strategy="most_frequent"),["fever"]),
    ("OneHot",OneHotEncoder(drop="first",sparse_output=False),["gender","city"]),
    ("Ordinal",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"])
    ],
    remainder="passthrough"
)
```

```python
x_train_new=transformer.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_new)
```
