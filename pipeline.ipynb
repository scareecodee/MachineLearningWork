```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# sperate input and output variables
x=df[["Pclass","Sex","Age","SibSp"	,"Parch"	,"Fare"	,"Embarked"]]
y=df[["Survived"]]
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
# imputation transfomer
tnf1=ColumnTransformer(
    transformers=[
    ("age_imputer",SimpleImputer(strategy="mean"),["Age"]),
    ("embarked_imputer",SimpleImputer(strategy="most_frequent"),["Embarked"])
    ],
    remainder="passthrough"
)
```

```python
# one hot encoding tranformer
from sklearn.preprocessing import OneHotEncoder
tnf2=ColumnTransformer(
    transformers=[
    ("sex_embarked_ohe",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),["Sex","Embarked"])
    ],
    remainder="passthrough"
)
```

```python
# scaling
from sklearn.preprocessing import MinMaxScaler
tnf3=ColumnTransformer(
    transformers=[
        ("scaler",MinMaxScaler(),["Age","Fare"])
    ],
    remainder="passthrough"
)
```

```python
#training model
from sklearn.tree import DecisionTreeClassifier
tnf4=DecisionTreeClassifier()
```

```python
#creating pipeline
from sklearn.pipeline import Pipeline,make_pipeline
pipe=Pipeline(
    steps=[
        ('tnf1',tnf1),
        ('tnf2',tnf2),
        ('tnf3',tnf3),
        ('tnf4',tnf4), # model training ---> so use fit to train model later

    ]
)
```

```python
#pipe.fit(x_train,y_train)
'''

NOTE 1 : Dont use ["Age","Embarked"] in column tranformer 
as after one hot encoding it wont work as ohe returns numpy array 

NOTE 2: dont create multiple tranformer as Your columns change position after every ColumnTransformer. 
meaning you cannt use indices as well like [2,6].

'''
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier

preprocessor = ColumnTransformer([
    ("age_imputer",
     SimpleImputer(strategy="mean"),
     ["Age"]),

    ("embarked",
     Pipeline([
         ("imputer", SimpleImputer(strategy="most_frequent")),
         ("encoder", OneHotEncoder(handle_unknown="ignore"))
     ]),
     ["Embarked"]),

    ("sex",
     OneHotEncoder(handle_unknown="ignore",sparse_output=False),
     ["Sex"])
], remainder="passthrough")

pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("model", DecisionTreeClassifier())
])

pipe.fit(x_train, y_train)
```

```python
pipe.score(x_test,y_test) # return for classifier --> accuracy score and for regressor r square
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.named_steps
```

```python
# cross validation
from sklearn.model_selection import cross_val_score
scores=cross_val_score(pipe,x_train,y_train,cv=5,scoring="accuracy")
```
