```python
import pandas as pd
```

```python
df=pd.read_csv("customer.csv")
```

```python
df.sample(n=10)
```

```python
# review , education ---> ordinal
# gender , purchased ---> nominal
```

```python
# seperate input and output variables 
x=df[["review","education"]] # not take first two column as here i am learning only ordinal encoding for now.
y=df[["purchased"]]
```

```python
x.sample()
```

```python
y.sample()
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# ordinal encoding ---> used for input variables
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
columns=x_train.columns
x_train=encoder.fit_transform(x_train)
x_test=encoder.transform(x_test)
```

```python
x_train=pd.DataFrame(data=x_train,columns=columns)
```

```python
# lable encoding --> used it for output variables
columns=y_train.columns
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```
