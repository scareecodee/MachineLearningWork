```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

1.Categorical Data

```python
#Countplot
sns.countplot(x=df["Survived"])
```

```python
import matplotlib.pyplot as plt
df["Survived"].value_counts().plot.pie(autopct='%1.1f%%',
    ylabel='') # use autopct for showing percentage
```

2. Numerical Data

```python

import matplotlib.pyplot as plt
```

```python
# Histogram
plt.hist(df["Age"],bins=5)
```

```python
#Distplot ---> probability
sns.distplot(df["Age"])
```

```python
# Boxplot
sns.boxplot(x=df["Age"])
```

```python
# skewness
df["Age"].skew()
```
