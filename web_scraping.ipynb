```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0 Safari/537.36",
    
    "Accept-Language": "en-US,en;q=0.9",

    "Accept": "text/html,application/xhtml+xml",
    
    "Referer": "https://google.com"
}

webpages = []

for i in range(1, 301):

    url = f"https://www.ambitionbox.com/list-of-companies?page={i}"

    response = requests.get(
        url,
        headers=headers,
        timeout=10
    )

    webpages.append(response.text)

    print("Done:", i)
```

```python
meta_info_list = []
rating_data = []

for webpage in webpages:

    soup = BeautifulSoup(webpage, "lxml")

    meta_info = soup.find("div",class_="companyCardWrapper__metaInformation")
    meta_info_list.append(meta_info)
```

```python
company_name=[]
ratings=[]
about=[]
for item in meta_info_list:
    company_name.append(item.find("h2").text.strip())
    ratings.append(item.find("div",class_="rating_text rating_text--md").text.strip())
    about.append(item.find("span",class_="companyCardWrapper__interLinking").text.strip())

```
