1. Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(x=tips["total_bill"],y=tips["tip"],hue=tips["sex"],style=tips["smoker"])
```

2. Barplot-(Categorical-Numerical)

```python
sns.barplot(x=tips["sex"],y=tips["total_bill"],hue=tips["smoker"])
```

3. Distplot

```python
sns.displot(titanic[titanic["survived"]==0]["age"]) # age of the peope those didnt survived 
```

```python
sns.displot(titanic[titanic["survived"]==1]["age"]) # age of the peope those  survive
```
